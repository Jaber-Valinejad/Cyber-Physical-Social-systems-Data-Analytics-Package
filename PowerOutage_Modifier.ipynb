# Houlry Outage +Dealing with missing values

```
Daily_dat=pd.DataFrame()
Daily_dat['time']=''
Daily_dat['outage']=''

for i in range(Data.shape[0]):
  for j in range(3):
     if  j==0:
         Daily_dat.at[3*i,'time']=Data['time'][i]
         Daily_dat.at[3*i,'outage']=Data['outage'][i]
     else:
         Daily_dat.at[3*i+j,'time']=Data['time'][i]+j
         Daily_dat.at[3*i+j,'outage']=np.NaN
        


#################################### time series Missing Value
print("Missing Value:",Daily_dat.isnull().sum())
Daily_dat['out']=pd.to_numeric(Daily_dat['outage'])
Daily_dat.drop(['outage'], axis=1, inplace=True)
# Interpolate missing value
Daily_dat['outage']=Daily_dat['out'].interpolate()
#Data['outage']=Data['outage'].interpolate(limit=1, limit_direction='forward')
# Forward-fill
#Data['outage']=Data['outage'].ffill()
# Back-fill
#Data['outage']=Data['outage'].bfill()
#Filna
#Data['outage']=Data['outage'].fillna(value=None, method='backfill', axis=None, limit=7, downcast=None)
Daily_dat.drop(['out'], axis=1, inplace=True)
Daily_dat.head() 
```
